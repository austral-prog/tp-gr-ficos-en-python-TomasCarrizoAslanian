## Grafico 1: Grafico de Barras

```python
plt.bar(['Matematica', 'Lengua', 'Ciencia'], [8, 3.5, 10])
plt.title("Notas de Materias")
plt.ylabel("Puntaje")
plt.savefig('imgs/graf_barras.png', dpi=300, bbox_inches='tight')
plt.show()
```

<p>Notas obtenidas en cada materia</p>

## Grafico 2: Grafico de Torta

```python
Materias = ['Analisis Matematico', 'Programacion', 'Filosofia', 'Introduccion']
horas = [12, 5, 2, 2]

plt.figure(figsize=(6, 6))
plt.pie(horas, labels=Materias, autopct='%1.1f%%', startangle=90)
plt.title('Porcentaje tiempo de materias')
plt.savefig('imgs/graf_torta.png', dpi=300, bbox_inches='tight')
plt.show()
```

<p>Porcentaje de horas semanales de cada materia del primer cuatrimestre</p>

## Grafico 3: Scatter Plot

```python
altura = [150, 160, 165, 170, 175, 180, 185]
peso =   [50,  60,  65,  68,  75,  80,  85]

plt.figure(figsize=(7, 5))
plt.scatter(altura, peso, color='green', marker='x')
plt.title('Relación entre altura y peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.savefig('imgs/graf_scatter.png', dpi=300, bbox_inches='tight')
plt.show()
```

<p>Relacion entre altura y peso de personas</p>

## Grafico 4: Grafico de Lineas

```python
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
Horas = [2, 3, 4, 0, 1, 4, 1]

plt.figure(figsize=(7, 4))
plt.plot(dias, Horas, marker='o', linestyle='-', color='orange')
plt.title('Tiempo de estudio por dia, en horas')
plt.xlabel('Día')
plt.ylabel('Horas')
plt.grid(True)
plt.savefig('imgs/graf_lineas.png', dpi=300, bbox_inches='tight')
plt.show()
```

<p>Horas de estudio dedicadas por dia en la semana</p>

## Grafico 5: Boxplot

```python
notas = [1, 2, 4, 2, 5, 4, 10, 8, 7, 5, 10, 10, 3, 7, 8]

plt.figure(figsize=(5, 4))
plt.boxplot(notas)
plt.title('Distribución de notas')
plt.ylabel('Nota')
plt.xticks([1], ['Notas'])
plt.grid(True)
plt.savefig('imgs/graf_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

<p>Distribucion de las notas obtenidas en parciales y trabajos</p>
